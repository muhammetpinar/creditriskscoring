```python
import pandas as pd 
import numpy as np 
 
import seaborn as sns 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('CreditScoring.csv') 
```

```python
df.head() 
```

```python
df.columns = df.columns.str.lower() 
```

```python
status_values = { 
    1: 'ok', 
    2: 'default', 
    0: 'unk' 
} 
```

```python
df.status = df.status.map(status_values) 
```

```python
home_values = { 
    1: 'rent', 
    2: 'owner', 
    3: 'private', 
    4: 'ignore', 
    5: 'parents', 
    6: 'other', 
    0: 'unk' 
} 
 
df.home = df.home.map(home_values)
```

```python
marital_values = { 
    1: 'single', 
    2: 'married', 
    3: 'widow', 
    4: 'separated', 
    5: 'divorced', 
    0: 'unk' 
} 
 
df.marital = df.marital.map(marital_values) 
 
 
records_values = { 
    1: 'no', 
    2: 'yes', 
    0: 'unk' 
} 
 
df.records = df.records.map(records_values) 
 
 
job_values = { 
    1: 'fixed', 
    2: 'partime', 
    3: 'freelance', 
    4: 'others', 
    0: 'unk' 
} 
 
df.job = df.job.map(job_values) 
```

```python
#degiskenlerdeki nümeriklere gerçek değerleri kategorikleri koyduk
```

```python
df.describe().round() 
```

```python
#max value deki 999999999 lar dikkate alıyorum
```

```python
for c in ['income', 'assets', 'debt']: 
    df[c] = df[c].replace(to_replace=99999999, value=np.nan)
```

```python
#nan yaptım
```

```python
df.describe().round()
```

```python
df.status.value_counts() 
```

```python
df = df[df.status != 'unk'] 
```

```python
#test train split
from sklearn.model_selection import train_test_split 
 
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=11) 
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=11)
```

```python
len(df_train), len(df_val), len(df_test) 
```

```python
#test val train sayıları
```

```python
y_train = (df_train.status == 'default').values 
y_val = (df_val.status == 'default').values
```

```python
del df_train['status'] 
del df_val['status'] #statusları veri setinden siliyorum
```

```python
df_train = df_train.fillna(0) 
df_val = df_val.fillna(0)  #na leri 0 ile doldurdum
```

```python
dict_train = df_train.to_dict(orient='records') 
dict_val = df_val.to_dict(orient='records')
```

```python
{'seniority': 10, 
 'home': 'owner', 
 'time': 36, 
 'age': 36, 
 'marital': 'married', 
 'records': 'no', 
 'job': 'freelance', 
 'expenses': 75, 
 'income': 0.0, 
 'assets': 10000.0, 
 'debt': 0.0, 
 'amount': 1000, 
 'price': 1400} 
```

```python
from sklearn.feature_extraction import DictVectorizer 
 
dv = DictVectorizer(sparse=False) 
 
X_train = dv.fit_transform(dict_train) 
X_val = dv.transform(dict_val) 
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
dt = DecisionTreeClassifier() 
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score  #performans metriği için area under the curve bakıyorum
```

```python
y_pred = dt.predict_proba(X_train)[:, 1] 
roc_auc_score(y_train, y_pred)      #1 olması suphelı val ıle bkaıyorum
```

```python
y_pred = dt.predict_proba(X_val)[:, 1] 
roc_auc_score(y_val, y_pred)         #overfit var
```

```python
dt = DecisionTreeClassifier(max_depth=2) 
dt.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text  
 
tree_text = export_text(dt, feature_names=dv.feature_names_)  
print(tree_text) 
```

```python
y_pred = dt.predict_proba(X_train)[:, 1] 
auc = roc_auc_score(y_train, y_pred)
print('train auc', auc) 
 
y_pred = dt.predict_proba(X_val)[:, 1] 
auc = roc_auc_score(y_val, y_pred) 
print('validation auc', auc)
```

```python
#önceki stepde aucları yaklastırdım ama çok sade model oldu depth parametresini optimize etmek gerek

for depth in [1, 2, 3, 4, 5, 6, 10, 15, 20, None]: 
    dt = DecisionTreeClassifier(max_depth=depth) 
    dt.fit(X_train, y_train) 
    y_pred = dt.predict_proba(X_val)[:, 1] 
    auc = roc_auc_score(y_val, y_pred) 
    print('%4s -> %.3f' % (depth, auc))
```

```python
#none da olabilecek max görünmekte 4 5 6 optimal gibi gözüküyor
```

```python
for m in [4, 5, 6]: 
    print('depth: %s' % m) 
 
    for s in [1, 5, 10, 15, 20, 50, 100, 200]: 
        dt = DecisionTreeClassifier(max_depth=m, min_samples_leaf=s) 
        dt.fit(X_train, y_train) 
        y_pred = dt.predict_proba(X_val)[:, 1] 
        auc = roc_auc_score(y_val, y_pred) 
        print('%s -> %.3f' % (s, auc)) 
     
    print()   #min leaf size'ı optimize ediyorum verdigi auclara göre
```

```python
#depth 6 ile mean leaf size' ı 15 olarak eğitmeye karar verdim
```

```python
dt = DecisionTreeClassifier(max_depth=6, min_samples_leaf=15) 
dt.fit(X_train, y_train) 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#10 agaçla eğiteceğim ilk olarak
rf = RandomForestClassifier(n_estimators=10, random_state=3) 
rf.fit(X_train, y_train) 
```

```python
y_pred = rf.predict_proba(X_val)[:, 1] 
roc_auc_score(y_val, y_pred) 
```

```python
aucs = [] #bos liste olusturdum auclar ıcın
 
for i in range(10, 201, 10): #agaç sayısını optimize edeceğim sayılar
    rf = RandomForestClassifier(n_estimators=i, random_state=3) #B 
    rf.fit(X_train, y_train) 
 
    y_pred = rf.predict_proba(X_val)[:, 1] 
    auc = roc_auc_score(y_val, y_pred) 
    print('%s -> %.3f' % (i, auc)) 
 
    aucs.append(auc) 
```

```python
plt.plot(range(10, 201, 10), aucs)  #ağaç sayılarına göre auclar
```

```python
#25 ile 75 arasında çok hızlı bir artış var daha sonra hız yavaşşlamış 
```

```python
#parametreleri toplu optimize edeceğim
```

```python
all_aucs = {} 
 
for depth in [5, 10, 20]: 
    print('depth: %s' % depth) 
    aucs = [] #C 
 
    for i in range(10, 201, 10): 
        rf = RandomForestClassifier(n_estimators=i, max_depth=depth, random_state=1) #D 
        rf.fit(X_train, y_train) 
        y_pred = rf.predict_proba(X_val)[:, 1] 
        auc = roc_auc_score(y_val, y_pred) 
        print('%s -> %.3f' % (i, auc)) 
        aucs.append(auc) 
     
    all_aucs[depth] = aucs 
    print() 
```

```python
#görselleştirelim
num_trees = list(range(10, 201, 10)) 
plt.plot(num_trees, all_aucs[5], label='depth=5') 
plt.plot(num_trees, all_aucs[10], label='depth=10') 
plt.plot(num_trees, all_aucs[20], label='depth=20') 
plt.legend() 
```

```python
#depth 10 okey min samples leaf a bakalım
```

```python
all_aucs = {} 
 
for m in [3, 5, 10]: 
    print('min_samples_leaf: %s' % m) 
    aucs = [] 
 
    for i in range(10, 201, 20): 
        rf = RandomForestClassifier(n_estimators=i, max_depth=10, min_samples_leaf=m, 
random_state=1) 
        rf.fit(X_train, y_train) 
        y_pred = rf.predict_proba(X_val)[:, 1] 
        auc = roc_auc_score(y_val, y_pred) 
        print('%s -> %.3f' % (i, auc)) 
        aucs.append(auc) 
     
    all_aucs[m] = aucs 
    print()
```

```python
num_trees = list(range(10, 201, 20)) 
plt.plot(num_trees, all_aucs[3], label='min_samples_leaf=3') 
plt.plot(num_trees, all_aucs[5], label='min_samples_leaf=5') 
plt.plot(num_trees, all_aucs[10], label='min_samples_leaf=10') 
plt.legend()
```
